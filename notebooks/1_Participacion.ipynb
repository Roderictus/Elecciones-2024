```python
# esto debe de agregar a secciones electorales
required_columns = ['ID_ENTIDAD', 'ENTIDAD', 'ID_DISTRITO_FEDERAL', 'DISTRITO_FEDERAL', 'SECCION']

vote_columns = ['PAN', 'PRI', 'PRD', 'PVEM', 'PT', 'MC', 'MORENA', 'PAN_PRI_PRD', 'PAN_PRD', 'PAN_PRI', 'PRI_PRD', 
                'PVEM_PT_MORENA', 'PVEM_PT', 'PVEM_MORENA', 'PT_MORENA', 'CANDIDATO/A NO REGISTRADO/A', 'VOTOS NULOS']


aggregation_columns = vote_columns + ['TOTAL_VOTOS_CALCULADOS', 'LISTA_NOMINAL']
# Ensure all vote columns and aggregation columns are numeric
pres[aggregation_columns] = pres[aggregation_columns].apply(pd.to_numeric, errors='coerce')

aggregated_data = pres.groupby(required_columns)[aggregation_columns].sum().reset_index()




# Aggregate vote columns by required columns
#aggregated_data = pres.groupby(required_columns)[vote_columns + ['TOTAL_VOTOS_CALCULADOS', 'LISTA_NOMINAL']].sum().reset_index()
```

```python

# Display the aggregated data
print(aggregated_data)

# 70504 Secciones
```

```python

# Calculate percentages
for col in vote_columns:
    aggregated_data[col + '_PERCENTAGE'] = (aggregated_data[col] / aggregated_data['TOTAL_VOTOS_CALCULADOS']) * 100

# Calculate participation
aggregated_data['PARTICIPATION_PERCENTAGE'] = (aggregated_data['TOTAL_VOTOS_CALCULADOS'] / aggregated_data['LISTA_NOMINAL']) * 100

# Display the final DataFrame
print(aggregated_data[["PAN_PERCENTAGE","PRI_PERCENTAGE",
                       "PRD_PERCENTAGE","MORENA_PERCENTAGE", "PT_PERCENTAGE","PAN_PRI_PRD_PERCENTAGE"]].head())

output_file_path = 'C:/Proyectos/Elecciones-2024/data/Agregados/Presidencial/TODO_pres_Seccion_aggregated.csv'  # Update with the desired file path
aggregated_data.to_csv(output_file_path, index=False)
```
