```python
CDMX = pres[pres["ENTIDAD"]  == "CIUDAD DE MÉXICO"]
```

```python
CDMX['ENTIDAD'].value_counts().sort_values()
```

```python
CDMX.columns
```

```python
required_columns = ['ID_ENTIDAD', 'ENTIDAD', 'ID_DISTRITO_FEDERAL', 'DISTRITO_FEDERAL', 'SECCION']

vote_columns = ['PAN', 'PRI', 'PRD', 'PVEM', 'PT', 'MC', 'MORENA', 'PAN_PRI_PRD', 'PAN_PRD', 'PAN_PRI', 'PRI_PRD', 
                'PVEM_PT_MORENA', 'PVEM_PT', 'PVEM_MORENA', 'PT_MORENA', 'CANDIDATO/A NO REGISTRADO/A', 'VOTOS NULOS']

# Aggregate vote columns by required columns
aggregated_data = CDMX.groupby(required_columns)[vote_columns + ['TOTAL_VOTOS_CALCULADOS', 'LISTA_NOMINAL']].sum().reset_index()

# Calculate percentages
for col in vote_columns:
    aggregated_data[col + '_PERCENTAGE'] = (aggregated_data[col] / aggregated_data['TOTAL_VOTOS_CALCULADOS']) * 100

# Calculate participation
aggregated_data['PARTICIPATION_PERCENTAGE'] = (aggregated_data['TOTAL_VOTOS_CALCULADOS'] / aggregated_data['LISTA_NOMINAL']) * 100

# Display the final DataFrame
print(aggregated_data[["PAN_PERCENTAGE","PRI_PERCENTAGE",
                       "PRD_PERCENTAGE","MORENA_PERCENTAGE", "PT_PERCENTAGE","PAN_PRI_PRD_PERCENTAGE"]].head())

output_file_path = 'C:/Proyectos/Elecciones-2024/CDMX_pres_aggregated.csv'  # Update with the desired file path
#aggregated_data.to_csv(output_file_path, index=False)


```
